# Importing important libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Data files into Python

```python
Global_Data = pd.read_csv('D:\Data Analyst NanoDegree\Explore Weather Trends\Global_Temp_Data.csv')
Local_Pune_Data = pd.read_csv('D:\Data Analyst NanoDegree\Explore Weather Trends\Pune_City_Temp_Data.csv')
```

# FInding NaN values

```python
Local_Pune_Data[Local_Pune_Data['avg_temp'].isnull()==True]
```

# Replacing NaN values with mean of avg_temp to make effect equal

```python
Local_Pune_Data['avg_temp'] = Local_Pune_Data['avg_temp'].fillna(np.mean(Local_Pune_Data['avg_temp']))
```

```python
Local_Pune_Data.head(20)
```

# Left Join data from Global_data with Local_Pune Data

```python
Total_Data = Local_Pune_Data.merge(Global_Data,how = 'left',on = 'year')

Total_Data
```

# Changing column names

```python
Total_Data = Total_Data.rename(columns = {'avg_temp_x' : 'Local_Temp',
                            'avg_temp_y' : 'Global_Temp'})

Total_Data
```

# Calculating Moving average for Local data and Global_data based on 10 years rolling period


```python
Total_Data['Local_Moving_Average'] = Total_Data['Local_Temp'].rolling(10).mean()

# Total_Data['Local_Moving_Average_MinPeriod'] = Total_Data['Local_Temp'].rolling(7,min_periods = 1).mean()



Total_Data['Global_Moving_Average'] = Total_Data['Global_Temp'].rolling(10).mean()


# Total_Data['Global_Moving_Average_Min_Period'] = Total_Data['Global_Temp'].rolling(7,min_periods = 1).mean()
```

# Adding column to take account difference between Lacal and Global temperatures.

```python
Total_Data['Difference'] = Total_Data['Local_Temp'] - Total_Data['Global_Temp']

Total_Data['Difference_Moving_Avg'] = Total_Data['Difference'].rolling(10).mean()
```

```python
Total_Data.tail(10)
```

# Setting graph size and grid style 

```python
plt.rcParams["figure.figsize"] = (20,10)
sns.set_style('white')
sns.set_context('poster')
```

```python
plt.plot(Total_Data['year'],Total_Data['Local_Moving_Average'],label = 'Local 10-year MA Temp.')
plt.plot(Total_Data['year'],Total_Data['Global_Moving_Average'], label = 'Global 10-year MA Temp.')
plt.plot(Total_Data['year'],Total_Data['Difference_Moving_Avg'], label = 'Difference in MA Temp.')


plt.xlabel('Year')
plt.ylabel('Temp. in c')
plt.title('Local Vs.Global Temp. Comparison')
plt.legend(loc="upper left", bbox_to_anchor= (1,1))
plt.annotate('8.551',xy = (1800,9))
plt.annotate('9.556', xy = (2013,10))
plt.annotate('25.446',xy = (2011,24))
plt.annotate('24.71', xy = (1800,24))
plt.annotate('16.165',xy = (1800,17))
plt.annotate('15.89', xy = (2011,17))
```
